### Correlations:

```python
# Create a correlation matrix using pandas

df_bank.corr()
```

```python
# Create a subplot with matplotlib

f,ax = plt.subplots(figsize=(15,15))
```

```python
# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.

corr = sns.heatmap(df_bank.corr(), annot = True, ax = ax)
```

It looks like that there are a lot of strong relationships between the economic type variables that contributes to a client's likeliness of subscribing to a term deposit.

#### Moderate Relationships

**num_of_employees/num_of_days_last_contact at 0.37**

**euribor_3_mths/num_of_days_last_contact at 0.3**

There is a some kind of positive indication between the number of employees and the number of days since last contacted with the client that would help the likliness of subscription.

#### Weak Relationships

Correlating the remaining variables will give either a weak positive relationship or weak to moderate negative relationships. Therefore this would mean they are is very little or no influence between these variables to contribute to the likeliness of a client subscribing to a term deposit.

### Categorical Plots:

```python
#Creates a histrogram for the "Euribor 3 Months Rate" variable
sns.distplot(df_bank['euribor_3_mths'], bins = 20)
```

```python
# Create a flag for the Euribor 3 Months rate variable to definite a rate category
df_bank.loc[df_bank['euribor_3_mths'] < 2, 'rate_category'] = 'Low Rate'
```

```python
df_bank.loc[(df_bank['euribor_3_mths'] >= 2) & (df_bank['euribor_3_mths'] < 4), 'rate_category'] = 'Middle Rate'
```

```python
df_bank.loc[df_bank['euribor_3_mths'] >= 4, 'rate_category'] = 'High Rate'
```

```python
df_bank['rate_category'].value_counts(dropna = False)
```

```python
df_bank.columns
```

```python
# Create a categorical plot for the "Euribor 3 Months Rate" with "Number of Employees" variable

sns.set(style="ticks")
catplot = sns.catplot(x="num_of_employees", y="euribor_3_mths", hue="rate_category", ax = ax, data=df_bank)
```

```python
# Create a categorical plot for the "Euribor 3 Months Rate" with "Employment Variation Rate" variable

sns.set(style="ticks")
catplot = sns.catplot(x="emp.var.rate", y="euribor_3_mths", hue="rate_category", ax = ax, data=df_bank)
```

It looks like that when the Employment Variation Rate and Number of Employees is higher, the Euribor 3 Months rate increases. This could be interepted that if the bank hires more quality employees, they are more likely to potentially get client's to subscribed to the term deposit due to the increase in the interest rate, which is the Euribor 3 Months Rate.
